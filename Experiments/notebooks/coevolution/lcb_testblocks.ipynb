```python
from common.coevolution import lcb_dataset
problems = lcb_dataset.load_code_generation_dataset(
    release_version="release_v6",
    start_date="2024-08-01",
    end_date="2025-05-10",
    difficulty=lcb_dataset.Difficulty.EASY,
)
```

```python
print(f"Loaded {len(problems)} problems.")


for problem in problems:
    test_type = problem.public_test_cases[0].testtype
    if test_type == lcb_dataset.TestType.STDIN:

        print(f"Problem ID: {problem.question_id}")
        # print(f"PROBLEM:\n{problem.question_content}")
        test_block = lcb_dataset.LCBDatasetTestBlockBuilder.build_test_class_block(problem.public_test_cases, problem.starter_code)

        print(test_block)
```
