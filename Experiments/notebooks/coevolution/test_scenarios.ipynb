# Scenairo based test case generation

```python
from common.coevolution import lcb_dataset
problems = lcb_dataset.load_code_generation_dataset(
    release_version="release_v5",
    start_date="2024-01-01",
    end_date="2024-12-31",
    difficulty=lcb_dataset.Difficulty.HARD,
)
```

```python
from common.llm_client import create_llm_client
from common.coevolution.prompt_templates import INITIAL_TEST_AGENT_CODER_STYLE as INITIAL_TEST
llm_client = create_llm_client(provider='openai', model='gpt-5-mini')


for problem in problems[:1]:
    print("========================================")
    print(f"Problem ID: {problem.question_id}")
    prompt= INITIAL_TEST.format(question_content=problem.question_content, starter_code=problem.starter_code, population_size=40)
    print(f"PROMPT:\n {prompt}\n")
    response = llm_client.generate(
        prompt=prompt
    )
    print(f"RESPONSE:\n{response}\n")   
```
