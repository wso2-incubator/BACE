```python
# CONSTANTS
Y = 0.1 # Probability that test cases are correct, resembles our prior belief for test correctness
X_0 = 0.6 # Initial belief that the code is correct, resembles our prior belief for code correctness
```

```python
import random

def get_X_n_when_true(X_prev: float, Y: float) -> float:
    """
    Computes the updated belief that the code is correct after a test case result.
    Args:
        X_prev: Previous belief that the code is correct (P(C))
        Y: Probability that the test case is correct (P(E1|C) and P(not E1|not C))
    """

    ratio = ((1-Y) / Y) * ((1 - X_prev) / X_prev)
    X_n = 1 / (1 + ratio)
    return X_n

def get_X_n_when_false(X_prev: float, Y: float) -> float:
    """
    Computes the updated belief that the code is correct after a test case result.
    Args:
        X_prev: Previous belief that the code is correct (P(C))
        Y: Probability that the test case is correct (P(E1|C) and P(not E1|not C))
    """

    ratio = (Y / (1 - Y)) * ((1 - X_prev) / X_prev)
    X_n = 1 / (1 + ratio)
    return X_n

print("Scenario 1: All tests pass")


print(f"X_0: {X_0}")
X_n = X_0
for i in range(1, 21):
    X_n = get_X_n_when_true(X_n, Y)
    print(f"X_{i}: {X_n}")


print("Scenario 2: All tests fail")
print(f"X_0: {X_0}")
X_n = X_0
for i in range(1, 21):
    X_n = get_X_n_when_false(X_n, Y)
    print(f"X_{i}: {X_n}")


print("\nScenario 3: Alternating test results (pass, fail, pass, fail, ...)")
X_n = X_0
print(f"X_0: {X_0}")
for i in range(1, 21):
    if i % 2 == 1:
        X_n = get_X_n_when_true(X_n, Y)
    else:
        X_n = get_X_n_when_false(X_n, Y)
    print(f"X_{i}: {X_n}")

print("\nScenario 4: Random test results")
random.seed(42)
X_n = X_0
print(f"X_0: {X_0}")
for i in range(1, 21):
    if random.choice([True, False]):
        X_n = get_X_n_when_true(X_n, Y)
    else:
        X_n = get_X_n_when_false(X_n, Y)
    print(f"X_{i}: {X_n}")
```
