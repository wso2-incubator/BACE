```python
from coevolution.adapters import lcb
problems = lcb.load_code_generation_dataset(
    release_version="release_v6",
    start_date="2025-03-01",
    end_date="2025-05-10",
    difficulty=lcb.Difficulty.HARD,
)
```

```python
problem_ids = [
    "arc195_b"
]


for problem_id in problem_ids:

    problem = next((p for p in problems if p.question_id == problem_id), None)
    if problem is None:
        print(f"Problem ID {problem_id} not found.")
        continue
    
    # print(f"Problem ID: {problem.question_id}")
    # print(f"Description: {problem.question_content}")
    print(problem.question_content, end="\n\n")
    print("You MUST follow the Starter Code format given below:")
    print(problem.starter_code, end="\n\n")


    # for i, test_case in enumerate(problem.private_test_cases):
    #     print(f"\n\n\nTest Case {i + 1}:")
    #     print(f"Input:\n{test_case.input}")
    #     print(f"Expected Output:\n{test_case.output}", end="\n\n")
```
